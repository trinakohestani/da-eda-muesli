# Merging datasets

Group order process data

```python
on_truck = order_process.groupby('order_id').agg({'order_date': 'min',
                                        'on_truck_scan_date': 'first',
                                        'ship_mode':'first'}).reset_index()
on_truck
```

Merge the grouped data to internal 

```python
ready_to_ship = on_truck.merge(internal, how='right')
ready_to_ship.drop('pickup_date', axis=1, inplace=True)
ready_to_ship
```

Merge the grouped data (on_truck) to campaign and name it arrival

```python
arrival = on_truck.merge(campaign, how='right')
arrival.drop('customer_name', axis=1, inplace=True)
arrival
```

## Change date type

Change date columns to datetime data and calculate number of days between each date in a new column

Business days between order and loading on truck

```python
import numpy as np
```

```python
# Change data to datetime data for 'order_date' and 'on_truck_scan_date'
on_truck['order_date'] = pd.to_datetime(on_truck['order_date'], dayfirst=True)
on_truck['on_truck_scan_date'] = pd.to_datetime(on_truck['on_truck_scan_date'], dayfirst=True)

#Calculate the difference in business days between 'order_date' and 'on_truck_scan_date'
on_truck['bus_days_between_order_and_truck'] = np.busday_count(on_truck['order_date'].dt.date.values.astype('datetime64[D]'), on_truck['on_truck_scan_date'].dt.date.values.astype('datetime64[D]'))

#Calculate the difference in calendar days between 'order_date' and 'on_truck_scan_date'
on_truck['cal_days_between_order_and_truck'] = (on_truck['on_truck_scan_date'] - on_truck['order_date']).dt.days
on_truck.head()
```

Days between order and ready to ship

```python
# Change data to datetime data for 'order_date' and 'ready_to_ship_date'
ready_to_ship['order_date'] = pd.to_datetime(ready_to_ship['order_date'], dayfirst=True)
ready_to_ship['ready_to_ship_date'] = pd.to_datetime(ready_to_ship['ready_to_ship_date'], dayfirst=True)

# Calculate the difference in business days between 'order_date' and 'ready_to_ship_date'
ready_to_ship['bus_days_between_order_and_ready'] = np.busday_count(ready_to_ship['order_date'].dt.date.values.astype('datetime64[D]'), ready_to_ship['ready_to_ship_date'].dt.date.values.astype('datetime64[D]'))

# Calculate the difference in calendar days between 'order_date' and 'ready_to_ship_date'
ready_to_ship['cal_days_between_order_and_ready'] = (ready_to_ship['ready_to_ship_date'] - ready_to_ship['order_date']).dt.days
ready_to_ship.head()
```

Days between ready to ship and loading on truck

```python
# Change data to datetime data for 'on_truck_scan_date'
ready_to_ship['on_truck_scan_date'] = pd.to_datetime(ready_to_ship['on_truck_scan_date'], dayfirst=True)

# Calculate the difference in business days between 'ready_to_ship_date' and 'on_truck_scan_date'
ready_to_ship['bus_days_between_ready_and_truck'] = np.busday_count(ready_to_ship['ready_to_ship_date'].dt.date.values.astype('datetime64[D]'), ready_to_ship['on_truck_scan_date'].dt.date.values.astype('datetime64[D]'))

# Calculate the difference in calendar days between 'ready_to_ship_date' and 'on_truck_scan_date'
ready_to_ship['cal_days_between_ready_and_truck'] = (ready_to_ship['on_truck_scan_date'] - ready_to_ship['ready_to_ship_date']).dt.days
ready_to_ship.head()
```

```python
# Change data to datetime data
arrival['order_date'] = pd.to_datetime(arrival['order_date'], dayfirst=True)
arrival['arrival_scan_date'] = pd.to_datetime(arrival['arrival_scan_date'], dayfirst=True)

# Calculate the difference in business days between 'order_date' and 'arrival_scan_date'
arrival['bus_days_between_order_and_arrival'] = np.busday_count(arrival['order_date'].dt.date.values.astype('datetime64[D]'), arrival['arrival_scan_date'].dt.date.values.astype('datetime64[D]'))

# Calculate the difference in calendar days between 'order_date' and 'arrival_scan_date'
arrival['cal_days_between_order_and_arrival'] = (arrival['arrival_scan_date'] - arrival['order_date']).dt.days
arrival
```
