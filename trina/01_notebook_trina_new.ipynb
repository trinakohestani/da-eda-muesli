# Merging datasets

Group order process data (needed to calculate the days between order placed and order loaded on truck)

```python
on_truck = order_process.groupby('order_id').agg({'order_date': 'min',
                                        'on_truck_scan_date': 'first',
                                        'ship_mode':'first'}).reset_index()
on_truck
```

Merge the grouped data to internal (needed to calculate the days between order placed and order ready to ship as well as the days between order ready to ship and order loaded on truck)

```python
ready_to_ship = on_truck.merge(internal, how='right')
ready_to_ship.drop('pickup_date', axis=1, inplace=True)
ready_to_ship
```

Merge the grouped data (on_truck) to campaign and name it arrival (needed to calculate the days between order placed and order delivered)

```python
arrival = on_truck.merge(campaign, how='right')
arrival.drop('customer_name', axis=1, inplace=True)
arrival
```

## Change date type

```python
import numpy as np
```

Change date columns to datetime data and calculate number of days between each date in a new column

Calculate business days between order and loading on truck

```python
# Change data to datetime data for 'order_date' and 'on_truck_scan_date'
on_truck['order_date'] = pd.to_datetime(on_truck['order_date'], dayfirst=True)
on_truck['on_truck_scan_date'] = pd.to_datetime(on_truck['on_truck_scan_date'], dayfirst=True)

#Calculate the difference in business days between 'order_date' and 'on_truck_scan_date'
on_truck['days_to_load_actual'] = np.busday_count(on_truck['order_date'].dt.date.values.astype('datetime64[D]'), on_truck['on_truck_scan_date'].dt.date.values.astype('datetime64[D]'))

on_truck.head()
```

Calculate business days between order placed and ready to ship

```python
# Change data to datetime data for 'order_date' and 'ready_to_ship_date'
ready_to_ship['order_date'] = pd.to_datetime(ready_to_ship['order_date'], dayfirst=True)
ready_to_ship['ready_to_ship_date'] = pd.to_datetime(ready_to_ship['ready_to_ship_date'], dayfirst=True)

# Calculate the difference in business days between 'order_date' and 'ready_to_ship_date'
ready_to_ship['days_to_ship_actual'] = np.busday_count(ready_to_ship['order_date'].dt.date.values.astype('datetime64[D]'), ready_to_ship['ready_to_ship_date'].dt.date.values.astype('datetime64[D]'))

ready_to_ship.head()
```

Calculate business days between ready to ship and loading on truck

```python
# Change data to datetime data for 'on_truck_scan_date'
ready_to_ship['on_truck_scan_date'] = pd.to_datetime(ready_to_ship['on_truck_scan_date'], dayfirst=True)

# Calculate the difference in business days between 'ready_to_ship_date' and 'on_truck_scan_date'
ready_to_ship['days_between_ready_and_truck_actual'] = np.busday_count(ready_to_ship['ready_to_ship_date'].dt.date.values.astype('datetime64[D]'), ready_to_ship['on_truck_scan_date'].dt.date.values.astype('datetime64[D]'))

ready_to_ship.head()
```

Calculate business days between order placed and order delivered

```python
# Change data to datetime data
arrival['order_date'] = pd.to_datetime(arrival['order_date'], dayfirst=True)
arrival['arrival_scan_date'] = pd.to_datetime(arrival['arrival_scan_date'], dayfirst=True)

# Calculate the difference in business days between 'order_date' and 'arrival_scan_date'
arrival['days_to_delivery_actual'] = np.busday_count(arrival['order_date'].dt.date.values.astype('datetime64[D]'), arrival['arrival_scan_date'].dt.date.values.astype('datetime64[D]'))

arrival.head()
```

## Hypothesis 1: Causes of Delays in Supply Chain
Most order delays can be explained by the time it takes for orders to be prepared for shipping and the time it takes for orders to be loaded onto the truck. Time for the shipping itself plays a subordinate role. (correlation analysis between these variables)

```python
# Define custom function for 75th percentile
def percentile_75(x):
    return np.percentile(x, 75)
```

### Delays Between Order Date and Loaded on Truck Date

```python
# Create new column with day of the week based of the order date
on_truck['order_day_of_week'] = on_truck['order_date'].dt.day_name()

# Create DataFrame for orders with 'Express' and for Standard Processing' shipping mode
express_on_truck = on_truck[on_truck['ship_mode'] == 'Express']
standard_on_truck = on_truck[on_truck['ship_mode'] == 'Standard Processing']

# Define dictionaries for standard and express shipping modes
standard_orderplaced_loadedontruck_busdays = {
    'Monday': 5,
    'Tuesday': 4,
    'Wednesday': 4,
    'Thursday': 6,
    'Friday': 4,
    'Saturday': 5,
    'Sunday': 5
}

express_orderplaced_loadedontruck_busdays = {
    'Monday': 3,
    'Tuesday': 4,
    'Wednesday': 3,
    'Thursday': 3,
    'Friday': 4,
    'Saturday': 3,
    'Sunday': 3
}

# Assign number of days for 'Standard Processing' shipping mode
standard_on_truck['days_to_load_target'] = standard_on_truck['order_day_of_week'].map(standard_orderplaced_loadedontruck_busdays)

# Assign number of days for 'Express' shipping mode
express_on_truck['days_to_load_target'] = express_on_truck['order_day_of_week'].map(express_orderplaced_loadedontruck_busdays)
```

Express

```python
express_on_truck.head()
```

```python
# Group by 'order_day_of_week' and aggregate
result = express_on_truck.groupby('order_day_of_week').agg({'days_to_load_actual': ['count', 'mean', percentile_75],
                                                            'days_to_load_target': ['count', 'mean', percentile_75]})

# Rename columns for clarity
result.columns = ['actual_count', 'actual_mean', 'actual_75th_percentile',
                  'target_count', 'target_mean', 'target_75th_percentile']

# Display the result
result
```

## On-Time Delivery Rate
An order was delivered on time when the 'days_to_delivery_actual' are equal to or smaller than the 'days_to_deliver_target'

```python
# Filter the dataset to include only on-time deliveries
on_time_deliveries_standard = standard_arrival[standard_arrival['days_to_delivery_actual'] <= standard_arrival['days_to_delivery_target']]

# Calculate the total number of orders
total_orders = len(standard_arrival)

# Calculate the number of on-time deliveries
on_time_count = len(on_time_deliveries_standard)

# Calculate the On-Time Delivery Rate
on_time_delivery_rate = on_time_count / total_orders * 100

# Print the result
print("On-Time Delivery Rate for Standard shipping: {:.2f}%".format(on_time_delivery_rate))
```

```python
# Filter the dataset to include only on-time deliveries
on_time_deliveries_express = express_arrival[express_arrival['days_to_delivery_actual'] <= express_arrival['days_to_delivery_target']]

# Calculate the total number of orders
total_orders = len(express_arrival)

# Calculate the number of on-time deliveries
on_time_count = len(on_time_deliveries_express)

# Calculate the On-Time Delivery Rate
on_time_delivery_rate = on_time_count / total_orders * 100

# Print the result
print("On-Time Delivery Rate for Express shipping: {:.2f}%".format(on_time_delivery_rate))
```
